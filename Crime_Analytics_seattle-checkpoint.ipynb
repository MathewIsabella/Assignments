```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from io import StringIO
import requests
```

```python
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/uwescience/datasci_course_materials/master/assignment6/seattle_incidents_summer_2014.csv'
```

```python
s=requests.get(url).text

crime=pd.read_csv(StringIO(s))
```

```python
crime.head()
```

```python
crime.info()
```

```python
crime["category"] = crime["Offense Type"].str.split("-").str[0]
```

```python
crime.info()
```

```python
import datetime
datadt = pd.DatetimeIndex(crime['Date Reported'])
crime['day_of_week'] = datadt.dayofweek
crime['date'] = datadt.date
crime['hour']= datadt.hour
crime['time']= datadt.time
crime['dow']= datadt.strftime('%a')
```

# PLOT 1: No:of crime Incidents reported Vs Hour

```python
sns.countplot(x='hour',data=crime)
plt.xlabel('Hour of occurance of crime')
plt.ylabel('No:of crime')
plt.title ('No:of crime vs hour')
```

From the above plot,it is inferred that the crime rate is highest around noon and at evening time.However, the type of crime occuring at the day/evening time needs to analysed.

# PLOT 2: Category of crime Incidents reported Vs Hour

```python
offenseType=crime.groupby(['hour','category'])
offenseType=offenseType.size().reset_index()
offenseType.columns= ['hour','category','count']
off=sns.FacetGrid(offenseType,hue='category',size=5,aspect=0.8)
off.map(plt.plot,'hour','count').add_legend()
plt.xlabel('Hour of occurance of crime')
plt.ylabel('Type of offense')
plt.title ('Offense type vs hour')
off.fig.autofmt_xdate()
```

From plot2 , it is understood that 'theft' is the category of crime that has maximum number of incidents reported.Theft, Burglary and vehicle theft are most common crime incident occuring during predominantly around noon and in the evening.

# PLOT 3: Category of crime Incidents reported Vs No:of Occurance

```python
plt.figure(figsize=(13,11))
crimeVal = crime.groupby("category")["RMS CDW ID"].describe().unstack().sort([('count')], ascending=False).index.values
sns.countplot(data=crime, x="category", order=crimeVal)
plt.xticks(rotation='vertical')
```

Plot 3 shows the count of crime incidents vs category of crime.The plot clearly helps in visualizing the top crime categories. Theft,Burglary,Vehicle theft,property damage,Asslt are the top 5 crime categories in Seattle.

# PLOT 4: No:of crime Incidents reported Vs District

```python
plt.figure(figsize=(13,11))
sns.countplot(data=crime, x="District/Sector",hue='dow',hue_order=('Mon','Tue','Wed','Thu','Fri','Sat','Sun'),palette='coolwarm')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

From the above plot, it is clear that most crime is reported in the District/Sectot 'M'and the no:of crime reported is highest on monday and least on sunday.The least no:of crimes are reported in district 99.

# PLOT 5: No:of crime Incidents reported Vs Days

```python
sns.countplot(x='dow',data=crime,order =('Sun','Mon','Tue','Wed','Thu','Fri','Sat'))
```

From the above plot, it is seen that most crime incidents are reported on monday and the least no:of incidents are reported on sunday.

# PLOT 6: No:of crime Incidents reported Vs Month

```python
sns.countplot(x='Month',data=crime,hue='dow',hue_order=('Mon','Tue','Wed','Thu','Fri','Sat','Sun'))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plt.figure(figsize=(20,20))
crime['hour8']=crime.groupby(crime['hour']<8)['hour'].transform('count')
sns.countplot(x='hour8',data=crime,hue='category',palette='coolwarm')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=8, fancybox=True, shadow=True)
```

```python
plt.figure(figsize=(20,20))
crime['hour13']=crime.groupby(crime['hour']==13)['hour'].transform('count')
sns.countplot(x='hour13',data=crime,hue='category',palette='coolwarm')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=8, fancybox=True, shadow=True)
```

# Crime_Analytics_SFO

```python
url1 = 'https://raw.githubusercontent.com/uwescience/datasci_course_materials/master/assignment6/sanfrancisco_incidents_summer_2014.csv'
```

```python
s=requests.get(url1).text

crimeSFO=pd.read_csv(StringIO(s))
```

```python
crimeSFO.head()
```

```python
crimeSFO.info()
```

# PLOT 1: No:of crime Incidents reported Vs Day

```python
sns.countplot(x='DayOfWeek',data=crimeSFO,order=('Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'))
```

In SFO area, it is seen that the no:of crime incidents is at peak on friday and gradually decreases over the weekends.

```python
datadt1 = pd.DatetimeIndex(crimeSFO['Date'])
crimeSFO['month'] = datadt1.month
crimeSFO["Timestamp"] = pd.to_datetime(crimeSFO.Date + " " + crimeSFO.Time)
crimeSFO["HourDay"] = crimeSFO.Timestamp.dt.hour
```

# PLOT 2: No:of crime Incidents reported Vs Month

```python
sns.countplot(x='month',data=crimeSFO,hue='DayOfWeek',hue_order=('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

In SFO, the highest no:of crime incidents are reported is in the month of August especially on friday and saturday

# PLOT 3: No:of crime Incidents reported Vs District

```python
plt.figure(figsize=(13,11))
sns.countplot(x='PdDistrict',data=crimeSFO,hue='DayOfWeek',hue_order=('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'))
plt.xticks(rotation='vertical')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

In SFO, most no:of crime incidents are reported in Southern district and least incidents are reported in Richmond district. More cops can be employed in Southern district area.

# PLOT 4b: No:of crime Incidents reported Vs Hour

```python
sns.countplot(x=crimeSFO.Timestamp.dt.hour, order=range(0,24))
plt.xticks(rotation='vertical')
```

The hourly plot clearly infers that the most crime incidents in SFO occur in the evening around 5 and 6 pm.

# PLOT 5b: No:of crime Incidents reported Vs Category

```python
crimeVal1 = crimeSFO.groupby("Category")["PdId"].describe().unstack().sort_values(by=["count"], ascending=False).index.values
sns.countplot(data=crimeSFO, x="Category", order=crimeVal1)
plt.xticks(rotation='vertical')
```

In SFO region, the category of crime reported highest is LATENCY/THEFT.

# PLOT 6b: No:of crime Incidents reported Vs hour

```python
offenseType1=crimeSFO.groupby(['HourDay','Category'])
offenseType1=offenseType1.size().reset_index()
offenseType.columns= ['HourDay','Category','count']
off=sns.FacetGrid(offenseType,hue='Category',size=5,aspect=0.8)
off.map(plt.plot,'HourDay','count').add_legend()
plt.xlabel('Hour of occurance of crime')
plt.ylabel('Type of offense')
plt.title ('Offense type vs hour')
off.fig.autofmt_xdate()
```

In SFO region, Theft is the most common crime incident reported which predominantly is high around the noon and in evening.
